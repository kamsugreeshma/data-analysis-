###pie chart for categoty count

```python
plt.pie(Categoty_Counts['count'],labels=Categoty_Counts['Category'])
```

#title

```python
plt.pie(Categoty_Counts['count'],labels=Categoty_Counts['Category'],autopct='%1.2f%%')
plt.title('distribution of categories by count')
```

#suprate chestusde

```python
plt.pie(Categoty_Counts['count'],labels=Categoty_Counts['Category'],autopct='%1.2f%%',explode=[0,0,0.05],)
plt.title('distribution of categories by count')
plt.axis('equal')
```

#colour

```python
plt.pie(Categoty_Counts['count'],labels=Categoty_Counts['Category'],autopct='%1.2f%%',explode=[0,0,0.05], colors=['red','pink','blue'])
plt.title('distribution of categories by count')
plt.axis('equal')
```

#angle

```python
plt.pie(Categoty_Counts['count'],labels=Categoty_Counts['Category'],autopct='%1.2f%%',explode=[0,0,0.05], colors=['red','pink','blue'],startangle=90)
plt.title('distribution of categories by count')
plt.axis('equal')
```

#shadow


```python
plt.pie(Categoty_Counts['count'],
labels=Categoty_Counts['Category'],
autopct='%1.2f%%',
explode=[0.1,0.1,0.05], 
colors=['red','pink','blue'],
startangle=90,
shadow=True)
plt.title('distribution of categories by count')
plt.axis('equal')
```

#colour describ chestude

```python
plt.pie(Categoty_Counts['count'],
labels=Categoty_Counts['Category'],
autopct='%1.2f%%',
explode=[0.1,0.1,0.05], 
colors=['red','pink','blue'],
startangle=90,
shadow=True)
plt.title('distribution of categories by count')
plt.axis('equal')
plt.legend()
```
